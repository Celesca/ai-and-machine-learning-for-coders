#### Import the datasets from TFDS to the format of ndarray like Keras.datasets
* batch_size = -1 meaning load in one batch.
* as_supervise meaning tuple in (input, label)

```python
(training_images, training_labels), (test_images, test_labels) = tfds.as_numpy(tfds.load('fashion_mnist',
                                                                                         split= ['train', 'test'],
                                                                                         batch_size = -1,
                                                                                         as_supervised = True))
```

```python
import tensorflow as tf

training_images = training_images / 255.0
test_images = test_images / 255.0

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28,1)),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.fit(training_images, training_labels, epochs=5)
```

### Image Augmentation with the TFDS
* Normally we need to declare the path directory to use ImageDataGenerator from Tensorflow

```python
data = tfds.load('horses_or_humans', split="train", as_supervised=True)

train_batches = data.shuffle(100).batch(10)
```

```python
# Mapping new images before shuffle and create batches

def augmentimages(image, label):
  image = tf.cast(image, tf.float32)
  image = (image/255)
  image = tf.image.random_flip_left_right(image)
  return image, label

# Map
train = data.map(augmentimages)

train_batches = data.shuffle(100).batch(32)
```
